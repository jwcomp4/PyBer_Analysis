# Annotate Charts

- Annotating charts tells the data story more clearly and in a more compelling manner.

## MATLAB Method

- **plt.figure(figsize=(w,h)**: Change the size of the figure in pixels. Added on the first line of the script
- **plt.plot(x, y, lable='line')**: Add a lable that will be added to the legend
- **plt.xlim(min, max)**: Set the min and max range of the x-axis. Can take strings or indexes
- **plt.ylim(min, max)**: Set the min and max range of the y-axis Can take strings or indexes
- **plt.xlabel('x label)**: Add a label to the x-axis, argument takes a string
- **plt.ylabel('y label)**: Add a label to the y-axis, argument takes a string
- **plt.title('Title')**: Add a title
- **plt.legend()**: Add a legend
- **plt.grid()**: Add a grid to the chart
- **plt.savefig("add a path and figure extension")**: Save the figure with the given extension Added at the end of the script

### Notes on Convention:
- always add labels to x and y axes
- if you have multiple graphs make each one stand out with different color use, line style, legend, etc.
- thoughtfully consider the ranges of your x and y axes.

### Documentation for reference:

- [Matplotlib doc on matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
- [Matplotlib doc on the Pyplot API](https://matplotlib.org/stable/api/index.html)
- [Matplotlib doc on anatomy of Matplotlib chart](https://matplotlib.org/stable/tutorials/introductory/usage.html)


```python
# Create a plot and add a label for the legend
# You can also add parameters to control the line width, markers, color, etc. within the plt.plot() 
# Do so in this order:
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#Creating labels for the x and y axes:
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45:
plt.ylim(0, 45)

# Create a title:
plt.title('PyBer Fare by Month')

# Adding a grid:
plt.grid()

# adding the legend:
plt.legend()
```

# Creating Bar Charts with MATLAB Approach

- Bar charts are helpful with discrete data (data based on counts) in distinct columns and for measuring values over time when line charts get too full.

- Can also create a bar chart to visualize data from 2 or more datasets, either by displaying them side by side or using a stacked chart.

- Bar charts are a good choice for time-series data or ordinal values (i.e. age ranges, salary ranges, or groups, etc.)

```python
# Creating a bar chart plot:
# Can annotate the chart as above, but don't need marker or linewidth attributes

plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create labels:
plt.xlabel('Date')
plt.ylabel('Fare($)')

#Create a title
plt.title('PyBer Fare by Month')

# Add a legend
plt.legend()
```

## Creating a Bubble Chart with MATLAB Method

- Bubble charts are helpful showing financial, population, weather data because you can add a third and fourth factor to convey more info.
     - The first 2 factors are x and y axes
     - Changing the size of the dot into a "bubble" is the third factor - **size**
     - If there are more than 1 dataset that uses the same axes, the fourth factor (**color**) can correspond to each dataset.
 
 - Making a bubble chart is essentially modifying the scatter plot.
     - Add the "s" (size) parameter to your plt.scatter(), and it must equal to a feature of the data.
     - In this example, you can make the size of the marker equal to data floating-point decimal values of the y-axis.
     - See below for an example:

```python
# creating a bubble chart by modifying the scatter code to include the "s" (size) parameter
# assigning s to the values in the y_axis list.

plt.scatter(x_axis, y_axis, s=y_axis)

# note that the bubbles are too small for to tell much information
```

```python
# adjust the size by multiplying the data in the y-axis by any value.
# here, we can iterate through the y_axis data, multiply it by 3 and append to a new list.
# note that initiating and populating a new list keeps your original data in tact.

#initiating the new empty list:
y_axis_larger =[]

# using a for-loop to populate the list and make larger markers:

for data in y_axis:
    y_axis_larger.append(data * 3)

# creating a new bubble chart with the larger markers:
# note that the y-axis is the original data from y_axis
# the s parameter is set to the larger values to depict the size better.
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

    - here is the format/syntax for *list comprehension*:
        - new_list = [expression for item in list if condition]
    - See example below:
    
### Documentation for reference:

- [Matplotlib doc for pyplot scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter)

- [Matplotlib doc for simple scatter plot](https://matplotlib.org/stable/gallery/shapes_and_collections/scatter.html#sphx-glr-gallery-shapes-and-collections-scatter-py)

```python
# creating the same bubble chart as above using list comprehension
# note that the list comprehension syntax goes inside the s parameter assignment:
# seems to work like a mini for-loop within the parameter assignment

plt.scatter(x_axis, y_axis, s=[i *3 for i in y_axis])
```

```python
# creating scatter plot with object-oriented method:

fig, ax = plt.subplots()

ax.scatter(x_axis, y_axis)
```

```python
# creating a bubble chart with the object-oriented method

fig, ax = plt.subplots()

ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# refactoring the above code for a different appearance using object-oriented method:
# markers to sky blue color
# marker size to 5x each data point
# color 20% transparent
# black edge to the circles
# linewidth of the circles 2 pts
# Add a legend for city of Boston
# Add a title : PyBer Fare by month
# switch the axes so Fare($) is on the x-axis
# invert y-axis so Jan is on the top.

fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c="xkcd:sky blue", linewidths=2, edgecolors="black", alpha=.8,
          label="Boston")

ax.set_xlabel("Fare($)")
ax.set_ylabel("Month")
ax.set_title("PyBer Fare by Month")
ax.legend()
plt.gca().invert_yaxis()
```

# Creating Pie Charts

- Allow you to depict percentage data. 
- Limited in that it can only depict one dataset.

## Using the MATLAB Method

- use the plt.pie() method
     - this method requires values that are in an array
     - can also add labels that are in an array for each pie (optional)
     
- When changing colors remember the following:
    - lighter, brighter, and analogous colors are a better choice over darker ones.
    - [Doc for CSS colors Matplotlib](https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py)
    
[Documentation for Object-Oriented Pie Chart Matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html)


```python
# Creating a pie chart with the MATLAB method
# first argument is the data (y_axis array) and setting the labels parameter to x_axis
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# can make this look better changing some more parameters
# explode parameter makes one of the wedges pop out, takes a tuple that must correspond to the data
# the value in the explode array is the fraction of the radius from the center of the pie for each wedge.
# you can also add percentages using the autopct parameter with formatting syntax
# the syntax .1f depict the percentage to 1 decimal place
# the % symbols formats the number to a percentage

# this code changes the size to make it look less croweded:

plt.subplots(figsize=(8, 8))

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct="%.1f%%")

# this code changes the size to make it look less croweded:

plt.subplots(figsize=(8, 8))
```

```python
# working on the colors:
# Assign 12 colors, one for each month

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct="%.1f%%")

plt.show()
```

```python
# Using the Object-Oriented Method
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(y_axis,
       labels=x_axis,
      explode=explode_values,
      colors=colors,
      autopct="%.1f%%", 
      shadow=True, 
      counterclock=False,
      startangle=90)



plt.show()
```
